```python
import pandas as pd

url = f'https://github.com/DataTalksClub/llm-zoomcamp/blob/main/04-monitoring/data/results-gpt4o-mini.csv?raw=1'
df = pd.read_csv(url)
```

```python
df = df.iloc[:300]
```

```python
df.head()
```

```python
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer('multi-qa-mpnet-base-dot-v1')
```

```python
answer_llm = df.iloc[0].answer_llm

embedded=embedding_model.encode(answer_llm)

embedded #answer -0.42
```

```python
import numpy as np
from tqdm.auto import tqdm
```

```python
answer_llm = df.iloc[0].answer_llm
answer_orig = df.iloc[0].answer_orig
answer_llm_v=embedding_model.encode(answer_llm)
answer_orig_v = embedding_model.encode(answer_orig)

answer_llm_v.dot(answer_orig_v)
```

```python
evaluations=[]

for i in tqdm(range(len(df))):
    answer_llm = df.iloc[i].answer_llm
    answer_orig = df.iloc[i].answer_orig
    answer_llm_v=embedding_model.encode(answer_llm)
    answer_orig_v = embedding_model.encode(answer_orig)
    dot_product=answer_llm_v.dot(answer_orig_v)
    evaluations.append(dot_product)

evaluations
    
```

```python
np.percentile(evaluations, 75)  #answer 31.67
```

```python
pd.DataFrame(evaluations).describe()
    
```

```python
evaluations_norm=[]

for i in tqdm(range(len(df))):
    answer_llm = df.iloc[i].answer_llm
    answer_orig = df.iloc[i].answer_orig
    answer_llm_v=embedding_model.encode(answer_llm)
    answer_orig_v = embedding_model.encode(answer_orig)
    norm1 = np.sqrt((answer_llm_v * answer_llm_v).sum())
    v_norm1 = answer_llm_v / norm1
    norm2 = np.sqrt((answer_orig_v * answer_orig_v).sum())
    v_norm2 = answer_orig_v / norm2



    dot_product=v_norm1.dot(v_norm2)
    evaluations_norm.append(dot_product)
```

```python
np.percentile(evaluations_norm, 75)  #answer 0.83
```
